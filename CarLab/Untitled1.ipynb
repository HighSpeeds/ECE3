```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
displacement={
    0:4, #dispacement to the left in centimeters
    1:3,
    2:2,
    3:1,
    4:0,
    5:-1,
    6:-2,
    7:-3,
    8:-4
}
```

```python
values=np.empty([8,9])

for i in displacement.keys():
    values[:,i]=np.mean(np.load(f"measurment{i}.npy"),axis=0)
```

```python
plt.plot(displacement.values(),values[0,:])
plt.plot(displacement.values(),values[1,:])
plt.plot(displacement.values(),values[3,:])
plt.plot(displacement.values(),values[7,:])
```
