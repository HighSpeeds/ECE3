```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.axvline(0.5)
plt.xticks(color='w')

plt.yticks(color='w')
plt.ylabel("I")
plt.xlabel("V")
plt.savefig("fig1")
```

```python
plt.axhline(0.5)
plt.xticks(color='w')

plt.yticks(color='w')
plt.ylabel("I")
plt.xlabel("V")
plt.savefig("fig2")
```

```python
V0=1
R=1
```

```python
V=np.arange(-1,1)
plt.plot(V,(V0-V)/R)
plt.xticks(color='w')

plt.yticks(color='w')
plt.ylabel("I")
plt.xlabel("V")
plt.savefig("fig3")
```

```python
I0=1
R=1
```

```python
V=np.arange(-1,1)
plt.plot(V,I0-(V)/R)
plt.xticks(color='w')

plt.yticks(color='w')
plt.ylabel("I")
plt.xlabel("V")
plt.savefig("fig4")
```
